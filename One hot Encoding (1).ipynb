# OneHotEncoding using Sklearn

```python
x = df.iloc[:,0:4]
```

```python
x.head()
```

```python
y = df.iloc[:,-1]
```

```python
y.head()
```

```python
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(drop="first",sparse_output=False,dtype = np.int32)
```

```python
x_new = ohe.fit_transform(x[["fuel","owner"]])

print(x_new)
```

```python
# converting to DataFrame
encoded_columns = ohe.get_feature_names_out(['fuel','owner'])
x_new_df = pd.DataFrame(x_new,columns = encoded_columns)

# Displaying the DataFrame
x_new_df.head()
```
