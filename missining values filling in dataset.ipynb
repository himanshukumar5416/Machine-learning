```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("C:\\Users\\himanshu\\Downloads\\train (1).csv")
df
```

```python
df.drop("Cabin",axis = 1,inplace=True)
df.head()
```

```python
# Advance data filling using Linear Regression
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
train = df[df["Age"].notnull()]
test = df[df["Age"].isnull()]
```

```python
train.head()
```

```python
test.head()
```

```python
# working on train dataset
y = train["Age"]
```

```python
x = train.drop(["Age","PassengerId","Name","Ticket"],axis =1)
```

```python
x.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
x["Sex"] = le.fit_transform(x["Sex"])
x["Embarked"]=le.fit_transform(x["Embarked"])
```

```python
x.head()
```

```python
lr.fit(x,y)
```

```python
# prediction of age in test data
test
```

```python
x_new = test.drop(["Age","PassengerId","Name","Ticket"],axis=1)
```

```python
x_new.head()
```

```python
x_new["Sex"]=le.fit_transform(x_new["Sex"])
x_new["Embarked"]=le.fit_transform(x_new["Embarked"])
```

```python

pred = lr.predict(x_new)
```

```python
pred 
```

```python
test["Age"]=pred
```

```python
test
```

```python
train
```

```python
final_data = pd.concat([train,test],axis=0)
```

```python
final_data.tail()
```

```python
final_data.reset_index(drop=True,inplace=True)
final_data.head()
```
